```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df
```

# TASK 1

```python
for i in range(len(df["FlightNumber"])):
    if i == 1 or i ==3:
        df["FlightNumber"].iloc[i]=df["FlightNumber"].iloc[i-1] + 10
        
df    
```

# TASK 2

```python
a=df["From_To"].apply(lambda x: x.split('_'))
From = []
To = []
for i in range(len(a)):
    From.append(a.iloc[i][0])
    To.append(a.iloc[i][1])
temp_df = pd.DataFrame({"From":From,"To":To})
temp_df
```

# TASK 3

```python
for i in range(len(temp_df)):
    temp_df["From"].iloc[i] = temp_df["From"].iloc[i][0].upper()+ temp_df["From"].iloc[i][1:].lower()
    temp_df["To"].iloc[i] = temp_df["To"].iloc[i][0].upper()+ temp_df["To"].iloc[i][1:].lower()

temp_df    
```

# TASK 4

```python
df.drop("From_To",axis = 1,inplace = True)
df = df.join(temp_df)
df
```

# TASK 5

```python
delay_1 = []
delay_2 = []
delay_3 = []
i = []
for i in df["RecentDelays"]:
    length = len(i)
    if length == 0:
        for j in range(0,3):
            i.append(np.nan)
    elif length == 1:
        for j in range(0,2):
            i.append(np.nan)
    elif length == 2:
        i.append(np.nan)
    else:
        pass
    
for i in df["RecentDelays"] :
    delay_1.append(i[0])
    delay_2.append(i[1])
    delay_3.append(i[2])
    
df["delay_1"] = delay_1
df["delay_2"] = delay_2
df["delay_3"] = delay_3

df.drop("RecentDelays",axis = 1,inplace = True)
df
```
